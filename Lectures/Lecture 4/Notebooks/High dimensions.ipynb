# Plots for dimensional curse

Show the volume of a hypersphere as a function of dimension, and compare this to a hypercube. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
```

```python
def Vsphere(D, r=1):
    """Calculate the volume of a hypersphere with dimension D"""
    
    d2 = D/2
    return np.pi**d2*r**D/gamma(d2+1)

def Rsphere(D, V=1):
    """Calculate the radius of a hypersphere with dimension D and volume V"""
    
    return (gamma(D/2+1)*V)**(1/D)/np.sqrt(np.pi)

def Vcube(D, r=1):
    """Calculate the volume of a cube with radius r
    
    Note that radius = 1/2 side length.
    """
    
    return (2*r)**D

def Vsphere_Vcube(D):
    return Vsphere(D)/Vcube(D)
```

```python
Ds = np.linspace(1, 15, 100)
Ds_int = np.linspace(1, 15, 15)
```

```python
plt.plot(Ds, Vsphere_Vcube(Ds))
plt.scatter(Ds_int, Vsphere_Vcube(Ds_int))
plt.xlabel("Dimension")
plt.ylabel(r"$V_{\rm sphere}/V_{\rm cube}$")
plt.savefig("Vsphere_to_cube.png")
```

```python
Ds = np.linspace(1, 30, 100)
Ds_int = np.linspace(1, 30, 30)
plt.plot(Ds, Vsphere(Ds))
plt.scatter(Ds_int, Vsphere(Ds_int))
plt.xlabel("Dimension")
plt.ylabel(r"$V_{\rm sphere}$")
plt.savefig("Vsphere.png")
```
